```python
# Creating a list
student_name=["Bipina", "Rayhan", "Aryan", "Mansa"]
print("List:", student_name)
```

```python
#Converting above list into tuple
student_tuple = tuple(student_name)
print("Tuple:", student_tuple)
```

```python
#Modifying List using append() method
student_name.append("Riya")  #append() method aids in adding the items/element at the end of the list 
print("Modified List:", student_name)
```

```python
#Modifying list using insert() method 
student_name.insert(0, "Prabin") #insert() method is used to add the element at the specific position
print("Modified List:", student_name)
```

```python
#Modifying list through index number by overwriting fourth item which falls under index 3 from student_name Aryan to Jiya
student_name[3]="Jiya"
print("Modified List:", student_name)
```

```python
# Modiying tuple datas isn't possible as memories are fixed so we can't perform insert() and append() method.
#But we can perform tuple slicing to get specific range of element
#and also negative indexing to access elements from the end of the tuple incase we don't know the exact size.
print("Negative indexed tuple elements:", student_name[-3:]) #extract last three elements
```

```python
print("Sliced tuple elements:", student_name[1: 3])
```

```python
#Program to remove duplicates from a list and print unique values
# Original list with duplicates
numbers = [10, 11, 12, 12, 13, 13, 14]

# Convert list to set to remove duplicates
unique_numbers = set(numbers)

# Convert back to list (optional)
unique_list = list(unique_numbers)

# Print result
print("Original List:", numbers)
print("Unique Values:", unique_list)
```

```python
# Original list of ages
ages = [18, 25, 30, 40, 50]

# Reversing the list using slicing
reversed_ages = ages[::-1]

# Printing both original and reversed age lists
print("Original Age List:", ages)
print("Reversed Age List:", reversed_ages)
```

```python
# Creating a dictionary of student details
student_detail = {
    "name": "Bipina",
    "age": 20,
    "course": "IT"
}
# Accessing the name
print("Student Name:", student_detail["name"])

# Removing age using pop()
student_detail.pop("age")

# Printing the updated dictionary
print("Updated Dictionary:", student_detail)

# Printing the length of dictionary
print("Length of Dictionary:", len(student_detail))
```

```python
# First taking input from user using input() method 
text = input("Enter a string: ")

# Checking if it contains only alphabets
if text.isalpha():
    print("Valid")
else:
    print("Invalid")
```
