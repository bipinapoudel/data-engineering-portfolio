```python
#Data cleaning -Replace null values with default value
data=[10, None, 20, None, 30]
cleaned_data = []

for d in data:
    if d is None:
        cleaned_data.append(0)   # default value
    else:
        cleaned_data.append(d)

print(cleaned_data)
```

Recursion- when a function calls itself

```python
#base case- condition to stop the recursion, recursive case-when function calling itself

#print numbers from 1 to 5

def print_number(n):
    if n==6: # base case
      return
    print(n)
    print_number(n+1)  #this is a recursive call
print_number(1)
```

Fibonacci series

Sequence where 0, 1, 2, 3, 5, 8, 13------
F(n)= f(n-1)+f(n-2)

```python
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)

print(fib(5))
```

```python
#f(1)=1, f(0)=0
```

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)

#print series
n=7
for i in range(n):
    print(fib(i))
```
