```python
#Encapsulation- one of the core concepts of Object-Oriented Programming (OOP). 
#It means bundling data (variables) and methods (functions) together inside a class, and restricting direct access to some of the object's components.
#Protects data from accidental modification
#Improves code security
#Makes code more maintainable and modular
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)

acc.deposit(500)
acc.withdraw(300)

print(acc.get_balance())   # ✅ Access through method
#print(acc.__balance)       # ❌ Error (private)
```

```python
# Modifying Private Data (Setter Method)
 
class employee:
    def __init__(self, salary):
        self.__salary = salary # 10000 - python does name mangling for private variables- emp._employee__salary = 10000
 
    #Setter Method - update salary safely
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary # 20000
 
    #Getter Method - used to access private data safely
    def get_salary(self):
        return self.__salary
 
 
emp = employee(10000)
# employee.__init(emp,10000)
 
emp.set_salary(20000) # internally - employee.set_salary(emp,20000)
print(emp.get_salary()) # employee.get_salary(emp)
#print(emp.__salary)
 
 
# Flow 1- salary = 10000 -> salary updated too 20000 -> output - 20000
```

```python
#Inheritance- one class (child/subclass) inherits properties and methods from another class (parent/superclass).

#It helps in code reuse, reducing duplication, and creating relationships between classes.
```

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

#Usage
d = Dog()

d.speak()   # inherited from Animal
d.bark()    # defined in Dog
```

```python
class Person:
    def __init__(self,name):
        self.name=name
    def show_name(self):
        print("Name: ", self.name)

class Employee(Person): #Inheriting Person
    def __init__(self, name, salary):
        self.name=name
        self.salary=salary

emp=Employee("Bipina", 10000)
emp.show_name() # Inherited method

print(emp.salary)
    
```

```python
class Person:
    def __init__(self,name):
        self.name=name
    def show_name(self):
        print("Name: ", self.name)

class Employee(Person): #Inheriting Person
    def __init__(self, name, salary):
        super().__init__(name) #calling parent constructor
        self.salary=salary
#super callMethod- call methods of the parent class, to initialize parent properties, to reuse parent logic

emp=Employee("Bipina", 10000)
#emp.show_name() # Inherited method

print(emp.salary)
print(emp.name)
```

Key Difference without vs with super()
without Super()-parent not initialized, missing data, error-prone
with Super()- parent initialized, complete object, correct approach

inheritance-reuse code from parent
super()- calling parent's setup

```python
#Polymorphism- Same method name, different behavior
#Makes code flexible and reusable
#Allows one interface to work with different data types
#Reduces complexity

#same action but different behaviour
#Two types:
#1. Method overriding- when child class provides its own implementation of a method already defined in parent class
```

```python
class Parent:
    def show(self):
        print("This is a person")

class Employee(Parent):
    def show(self): #Overriding parent method
        super().show() #call parent method
        print("This is an employee")

obj=Employee()
obj.show()  #child method override the parent class
```

```python
#Method overloading-handle different inputs, same function name but have different number or types of arguments
def add (a, b = 0, c = 0):
    return a+b+c
 
#same function behaves differentl based on inputs
print(add(10))
print(add(10,20))
print(add(10,20,30))
```

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

#Usage
animals = [Dog(), Cat()]

for animal in animals:
    animal.speak()
```

```python
#Abstraction- hiding implementation details and showing only the essential features of an object.
#Reduces complexity--hiding complexity
#Improves code readability
#Enhances security (hides internal details)
```

```python
#From abstract class
from abc import ABC, abstractmethod

class Vehicle(ABC):   # Abstract class
    
    @abstractmethod
    def start(self):
        pass

#Child class implementation
class Car(Vehicle):
    def start(self):
        print("Car starts with key")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with kick")

#Usage
c = Car()
c.start()

b = Bike()
b.start()
```
