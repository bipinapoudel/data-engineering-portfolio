```python
#create data frame and add revenue to it
import pandas as pd
 
data1 = {
    "Product" : ["Phone", "Tablet", "Micro", "Keyboard"],
    "Price" : [50000, 60000, 20000, 50000],
    "Sales" : [10,20,15, 25]
}
df=pd.DataFrame(data1)
print(df)
```

```python
#adding a new column revenue
df["Revenue"]=df["Price"] * df["Sales"]
print(df)
```

```python
#add a record to the data
#add new record using LOC - to locate row
#to return one or more specified rows
df.loc[len(df)]=["Mouse", 2000, 33, 2000*33]
print(df)
```

```python
#to see index value 0 and 1
print(df.loc[[2, 3]])
```

```python
#Missing values
df.loc[2, "Sales"]=None
print(df)
```

```python
#print(df.columns)
df.drop(["sales", "MRV"], axis=1, inplace=True, errors="ignore")
print(df)
```

```python
df["Sales"].fillna(0, inplace=True)
print(df)
```

```python
filtered_df=df[df["Revenue"]>550000]
#Sorting revenue by ASC
result=filtered_df.sort_values(by="Revenue", ascending=True)
print(result)
```

```python
df.groupby("Product").sum()
print(df)
```

```python
df.loc[len(df)]=["Mouse", 2000, 20, 2000*20]
print(df)
```

```python
#Remove duplicate product
df = df.drop_duplicates(subset="Product", keep="last")
print(df)
```

```python
# Remove rows where Sales = 0
df=df[df["Sales"]!=0]
print(df)
```

```python
#Count the missing value
print(df.isnull().sum())
```

```python
df.loc[len(df)]=["A", 1400, 30, None]
df["Revenue"]=df["Price"]*df["Sales"]
print(df)
```

```python
#Try catch to handle exception

#Problem: Position-based assignment

#we are giving a list, which means:

#Pandas assigns values by column order

#So it expects:

#(Product, Price, Sales, Revenue)
try:
    df.loc[len(df)] = ["B", 1400, 30]   # may cause error if columns mismatch
    df["Revenue"] = df["Price"] * df["Sales"]
    print(df)

except ValueError as e:
    print("ValueError occurred:", e)

except KeyError as e:
    print("Column not found:", e)

except Exception as e:
    print("Some other error occurred:", e)
```

```python
#Try catch with dictionary

try:
    df.loc[len(df)] = {
        "Product": "C",
        "Price": 1400,
        "Sales": 30
    }

    df["Revenue"] = df["Price"] * df["Sales"]
    print(df)

except Exception as e:
    print("Error:", e)
```
