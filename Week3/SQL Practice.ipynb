## explain difference between count(*), count(1), and count(column)

count(*)-count all rows, includes null
count(1)-count rows, same as count(*)
count(column)- count non-null, ignore null

#Find highest salary per department and sorted

```python
%sql
SELECT department, MAX(salary) AS highest_salary FROM emp
GROUP BY department
ORDER BY highest_salary DESC;
```

```python
%sql
SELECT 
department,
AVG(salary) AS average_salary
FROM emp
GROUP BY department
ORDER BY average_salary DESC;
```

```python
%sql
SELECT emp_id, name, department, salary
FROM (
    SELECT *,
           DENSE_RANK() OVER (
               PARTITION BY department 
               ORDER BY salary DESC
           ) AS rank
    FROM employees
) t
WHERE rank <= 3
ORDER BY department, salary DESC;
```

```python
%sql
 
--filter top 3
-- subquery

-- inner query - assigns rank per department
-- outer query - filters only top 3, rank <=3


select *
from(
  select *,
  row_number() over(partition by department order by salary DESC) as rank
  from employees
)temp
where rank <=3;
```

INNER JOIN
Get employees with their department names -only matching records

```python
%sql
 --Inner Join->order table+customer table-get order with customer details
 --Left join- banking-> accounts+transactions-> show all accounts even if no transaction
 --Left join + WHERE condition - can behave like Inner join
 --Full outer join- HR analytics-> emp+dept-> identify missing mapping
 
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
INNER JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql

SELECT e.emp_id, e.emp_name, d.department_name
from employee e
LEFT JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
RIGHT JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql
--Full outer join combination of left and right join
--get all employees from all departments

SELECT e.emp_id, e.emp_name, d.department_name
from employee e
FULL JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql
--CROSS JOIN - Catresian product

SELECT e.emp_name, d.department_name
from employee e
CROSS JOIN departments d;
```

-- Row_number - Unique rank(no duplicates)
-- Rank() - skips rank() - 1,2,2,4
-- DENSE_RANK() - No skipping - no skipping (1,2,2,3)
 
-- Row_number() - unique, consective, numbering (pagination) - assigning unique IDs, or picking the top 1 item when ties doesn't matter
 
-- rank() - to rank items where ties skip intermediate rank (1,2,2,4)
-- assign the same rank to ties but leaves a gap in sequence e.g., sports ranking, where tied scores skips the next ranking position
 
-- dense_rank() - for ranking without gaps (1,2,2,3) -
-- assign the same rank to ties without gaps (1,2,2,3)
-- Leaderboards -> where ranking must be contiguous, identifying top N products
 
 
e.g.,
     rn
100  1  1  1
100  2  1  1
80   3  3  2
70   4  4  3
