CASE STUDY 1: 

```sql
/* 
1: Daily Active Users (DAU)
You have a table name – user_activity and you have columns user_id and activity_date
Table structure:
(1, '2024-01-01'),
(2, '2024-01-01'),
(1, '2024-01-02'),
(3, '2024-01-02'),
(2, '2024-01-02')
Problem -  Find number of active users per day
*/
```

```sql
%sql
--Creating Table
CREATE TABLE IF NOT EXISTS user_activity (
    user_id INT,
    activity_date DATE
);
```

```sql
%sql
--Inserting values in the table
INSERT INTO user_activity VALUES
(1, '2024-01-01'),
(2, '2024-01-01'),
(1, '2024-01-02'),
(3, '2024-01-02'),
(2, '2024-01-02');
```

```sql
%sql
--Finding number of active users per day
SELECT activity_date, COUNT(DISTINCT user_id) AS active_users_per_day 
FROM user_activity
GROUP BY activity_date
ORDER BY activity_date;
```

```sql
/*
2: Retention (Next Day Login)
Problem – Find users who logged in on consecutive days
*/
```

```sql
%sql
--Finding users who logged in on consecutive days
SELECT DISTINCT a.user_id
FROM user_activity a
JOIN user_activity b
ON a.user_id = b.user_id
AND b.activity_date = DATE_ADD(a.activity_date, 1);
```

CASE STUDY 2: 

```sql
/*
1: Top Product per Category
You have a table name is sales with column name category, product and revenue
Table structure:
('Electronics', 'Phone', 1000),
('Electronics', 'Laptop', 2000),
('Electronics', 'Tablet', 1500),
('Clothing', 'Shirt', 500),
('Clothing', 'Jacket', 800)
Problem – Find highest selling product in each category.
*/
```

```sql
%sql
--Creating sales table
CREATE TABLE IF NOT EXISTS sales (
  category VARCHAR(50),
  product VARCHAR(50),
  revenue INT
);
```

```sql
%sql
--Inserting values into sales table
INSERT INTO sales VALUES
('Electronics', 'Phone', 1000),
('Electronics', 'Laptop', 2000),
('Electronics', 'Tablet', 1500),
('Clothing', 'Shirt', 500),
('Clothing', 'Jacket', 800);
```

```sql
%sql
--Finding highest selling product per category
SELECT *
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY category ORDER BY revenue DESC) AS rnk
    FROM sales
) t
WHERE rnk = 1;
```

CASE STUDY 3: 

```sql
/*
1: Latest Transaction per User
You have table transactions with column name – user_id(int), txn_date (date) and amount(int)
*/
```

```sql
%sql
--Creating transactions table
CREATE TABLE IF NOT EXISTS transactions (
  user_id INT,
  txn_date DATE,
  amount INT
);
```

```sql
%sql
--Inserting values into transactions table
INSERT INTO transactions VALUES
(1, '2024-01-01', 100),
(1, '2024-02-01', 200),
(2, '2024-01-05', 300);
```

```sql
%sql
--Finding latest transaction for each user
SELECT user_id, txn_date, amount
FROM (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY txn_date DESC) AS latest_rank
    FROM transactions
) t
WHERE latest_rank = 1;
```

CASE STUDY 4: 

```sql
/*
Funnel Analysis
You have table funnel with column names – user_id and step (e.g., view, cart and purchase)
*/
```

```sql
%sql
--Creating funnel table
CREATE TABLE IF NOT EXISTS funnel (
    user_id INT,
    step VARCHAR(20)
);
```

```sql
%sql
--Inserting values into funnel table
INSERT INTO funnel VALUES
(1, 'view'),
(1, 'cart'),
(1, 'purchase'),
(2, 'view'),
(2, 'cart');
```

```sql
%sql
--Counting users who completed all steps (view -> cart -> purchase)
SELECT COUNT(DISTINCT user_id) AS steps_completed_users
FROM (
    SELECT user_id
    FROM funnel
    WHERE step IN ('view', 'cart', 'purchase')
    GROUP BY user_id
    HAVING COUNT(DISTINCT step) = 3
) t;
```
