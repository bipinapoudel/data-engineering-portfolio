```python
#Creating raw data
data=[
  (1, "Laptop", 1000),
  (2, "Tablet", 500),
  (3, "Phone", 3000)
]
#Defining columns
columns=["order_id", "Products","Price"]
df=spark.createDataFrame(data, columns)
#display(df)

#Apply filter and select
result=df.filter(df.Price>500).select("Products", "Price")
display(result)
#shows table format by filter
```

```python
#Grouped by Aggregation
#Find total sales by product
from pyspark.sql.functions import sum
data1=[
  (1, "Laptop", 1000),
  (2, "Tablet", 500),
  (3, "Phone", 3000),
  (4, "Phone", 4000)
]
columns1=["order_id1", "Products1","Price1"]
df1=spark.createDataFrame(data1, columns1)
#Group by Products and sum Price
result1=df1.groupBy("Products1").agg(sum("Price1").alias("Total_sales_per_product"))
display(result1)
```

```python
%python
#sorting data-show all orders by price
result = df1.orderBy(df1.Price1.desc())
display(result)
```

```python
#Creating any new column-add a new column discount_price=price-100
from pyspark.sql.functions import *
df=spark.createDataFrame(data1, columns1)
new_result=df.withColumn("discount_price", col("Price1")-100)
display(new_result)
```
