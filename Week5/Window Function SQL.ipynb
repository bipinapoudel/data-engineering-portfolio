```sql
--Aggregate Function 
select max(salary) as max_salary from employee1;
```

```sql
--finding max salary earned by employee in each department

select department, max(salary) as max_salary from employee1 group by department;
```

```sql
--trying to find out the max salary earned by employee in each department along with the details of the employee
select e.*, max(salary) over() as max_salary from employee1 e;
```

```sql
select e.*, max(salary) over(partition by department) as max_salary from employee1 e;
```

```sql
select * from(
select e.*, row_number() over(partition by department order by salary desc) as rn from employee1 e)x
where x.rn=1
;
```

```sql
--finding first 2 employees who joins each department 
select * from (
    select e.*, row_number() over (partition by department order by hire_date) as rn from employee1 e
)x
where x.rn<3;
```

```sql
--in case there is no any hire_date
select * from (
    select e.*, row_number() over (partition by department order by emp_id) as rn from employee1 e
)x
where x.rn<3;
```

```sql
--Rank() will skip a value for every duplicate value it found previously e.g--1,2,2,4,5,5,7
--dense_rank() will not skip any value for every duplicate value it found previously e.g--1,2,2,3,4,4,5
```

```sql
--finding top 2 employees in each department earning the max salary
select * from(
select e.*, rank() over(partition by department order by salary desc) as rnk from employee1 e) x
where x.rnk<3;
```

```sql
--finding top 2 employees in each department earning the max salary
select * from(
select e.*,dense_rank() over(partition by department order by salary desc) as dense_rnk from employee1 e) x
where x.dense_rnk<3;
```

```sql
--Using a sub query to find the max salary of the employee in each department with all the details of the employee
select e.* 
from employee1 e 
inner join ( select department, max(salary) as max_salary from employee1 group by department) x
on e.department=x.department
and e.salary=x.max_salary;
--here first finding max salary per department and then gets full employee details
```

```sql
--We can't approach the same pattern for finding the two max salary following the same pattern of sub query.
SELECT e.*
FROM employee1 e
JOIN (
    SELECT department, MAX(salary) AS max_salary
    FROM employee1
    GROUP BY department
    order by salary desc
    where limit 2
) m
ON e.department = m.department
AND e.salary = m.max_salary;
```

```sql
--It will fetch all top 3 salaries of employee in an employee table irrespective of department
select * from employee1 order by salary desc limit 3;
```
