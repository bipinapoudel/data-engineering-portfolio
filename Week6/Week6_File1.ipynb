```sql
%python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

# Employee data
data_emp = [
    (1, "John", "Doe", 60000, 1),
    (2, "Jane", "Smith", 80000, 2),
    (3, "Sam", "Brown", 75000, 1),
    (4, "Lisa", "Ray", 90000, 2),
    (5, "Tom", "Hanks", 50000, 1)
]

columns_emp = ["id", "first_name", "last_name", "salary", "department_id"]

df_emp = spark.createDataFrame(data_emp, columns_emp)

# Department data
data_dept = [
    (1, "marketing"),
    (2, "engineering")
]

columns_dept = ["id", "department"]

df_dept = spark.createDataFrame(data_dept, columns_dept)
```

```sql
%python
df_emp.display()
df_dept.display()
```

```sql
%python
df_join = df_emp.join(df_dept, df_emp.department_id == df_dept.id)
```

```sql
%python
from pyspark.sql.functions import max, when, abs

result = df_join.select(
    abs(
        max(when(df_join.department == "marketing", df_join.salary)) -
        max(when(df_join.department == "engineering", df_join.salary))
    ).alias("salary_difference")
)
result.show()
```
